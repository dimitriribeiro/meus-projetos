# Objetivo

Quando criamos um formulario e estamos trabalhando com os inputs de usuarios, muitas vezes lidamos com erros de digitação. Pensando nesse contexto, foi criado esse código para criar uma lista de caracteres embaralhar em minúsculas e maiúsculas. Dessa forma, se o input de nome vir no formato 'gaTo', podemos criar uma condição para saber se esse erro está contido na lista e retornar a correção utlizando métodos de string como .upper(), .lower() e .capitalize()

# Bibliotecas utilizadas

```python
import secrets
```

# Classe Embaralhador

```python
class Embaralhador():

  def __init__(self):
    self.palavra = None
    self.letras_minusculas = []
    self.letras_maiusculas = [] 
    self.novas_palavras = []
    self.conjuntos = []

  def embaralhar(self, palavra):

    if len(palavra) <= 1:
      return 'Você digitou uma letra e não uma palavra!'

    for letra in palavra:
      self.letras_minusculas.append(letra.lower())
      self.letras_maiusculas.append(letra.upper())

    for i in range(len(palavra)):
      conjunto = [self.letras_maiusculas[i]] + [self.letras_minusculas[i]]
      self.conjuntos.append(conjunto)

    for i in range(2**len(palavra)):
      contador = 0
      while contador < len(palavra)+1:
        nova_palavra = ''
        for x in self.conjuntos:
          nova_palavra += secrets.choice(x)
        contador += 1
        self.novas_palavras.append(nova_palavra)
    d = list(set(self.novas_palavras))
    return list(set(self.novas_palavras))
  def __len__(self):
    return self.d
    
if __name__ == '__main__':
  conjunto = Embaralhador()
  conjunto = conjunto.embaralhar('yuri')
  tamanho = len(conjunto)
  print(tamanho, conjunto)
```
