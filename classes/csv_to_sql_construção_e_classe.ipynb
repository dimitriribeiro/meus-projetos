- Formando a sintaxe SQL

```python
lista_sql_create = ['CREATE TABLE ', f'{csv.name}'.replace('.csv', ''), '(', '\n']
lista_sql_insert = ['INSERT INTO ', f'{csv.name}'.replace('.csv', ''), '(', '\n']
tipos_sql_dados = ['INT','DOUBLE PRECISION', 'DOUBLE PRECISION', 'VARCHAR', 'VARCHAR'] # tem que ser feita coloca na função da classe
sizes = [10,0,0,50,2] # tem que ser feita coloca na função da classe

for size in sizes:
  if size == 0:
    indice = sizes.index(size)
    sizes.remove(size)
    sizes.insert(indice, '')


for i,linha in enumerate(arquivo_sql):
  if i == 0:
    indice_para_size_tipos = 0

    for elemento in linha.split(','):
      
      if elemento == linha.split(',')[-1]:
        if sizes[indice_para_size_tipos] == '':
          lista_sql_create.append('\t'+elemento.replace('\n', '')+ f' {tipos_sql_dados[indice_para_size_tipos]}{sizes[indice_para_size_tipos]}')
          lista_sql_insert.append('\t'+elemento)
        else:
          lista_sql_create.append('\t'+elemento.replace('\n', '')+ f' {tipos_sql_dados[indice_para_size_tipos]}({sizes[indice_para_size_tipos]})')
          lista_sql_insert.append('\t'+elemento)

      else:
        if sizes[indice_para_size_tipos] == '':
          lista_sql_create.append('\t'+elemento.replace('\n', '')+ f' {tipos_sql_dados[indice_para_size_tipos]}{sizes[indice_para_size_tipos]}'+',\n')
          lista_sql_insert.append('\t'+elemento+','+'\n')
        else:
          lista_sql_create.append('\t'+elemento.replace('\n', '')+ f' {tipos_sql_dados[indice_para_size_tipos]}({sizes[indice_para_size_tipos]})'+',\n')
          lista_sql_insert.append('\t'+elemento+','+'\n')
      indice_para_size_tipos += 1

    lista_sql_create.append(');\n')
    lista_sql_insert.append(') VALUES ')
  else:
    lista_sql_insert.append('(')
    for dado in linha.split(','):
      if '"' in dado:
        dado = dado.replace('"', "'")
        lista_sql_insert.append(f"{dado}"+',')
      elif '.' in dado and '-' in dado:
        dado = float(dado)
        lista_sql_insert.append(f"{dado}"+',')
      else:
        lista_sql_insert.append(f"'{dado}'"+',')
    lista_sql_insert.append('),')

sql_create = ''.join(lista_sql_create)
sql_insert = ''.join(lista_sql_insert).replace(',)', ')')[:-1]
```

- Teste piloto

```python
print(sql_create.replace('"', ''))
print(sql_insert.replace('"', ''))
```

- Baixando o arquivo em formato .SQL

```python
arquivo_sql = open('nome_arquivo_final.sql', 'w')
arquivo_sql.write(sql_create)
arquivo_sql.write(sql_insert)
arquivo_sql.write(';')
arquivo_sql.close()
```

```python
#aplicando em uma classe
class SqlToCsv:
    def __init__(self):
      pass
    def to_sql(self, caminho_arquivo_csv, tipos, tamanhos, nome_arquivo):
      self.caminho_arquivo_csv = caminho_arquivo_csv
      self.tipos = tipos
      self.tamanhos = tamanhos
      self.nome_arquivo = nome_arquivo
      lista_sql_create = ['CREATE TABLE ', f'{self.nome_arquivo}', '(', '\n']
      lista_sql_insert = ['INSERT INTO ',f'{self.nome_arquivo}', '(', '\n']

      for size in self.tamanhos:
        if size == 0:
          indice = self.tamanhos.index(size)
          self.tamanhos.remove(size)
          self.tamanhos.insert(indice, '')

      with open(self.caminho_arquivo_csv, 'r') as csv:
        arquivo_sql= csv.readlines()

      for i,linha in enumerate(arquivo_sql):
        if i == 0:
          indice_para_size_tipos = 0

          for elemento in linha.split(','):
            
            if elemento == linha.split(',')[-1]:
              if self.tamanhos[indice_para_size_tipos] == '':
                lista_sql_create.append('\t'+elemento.replace('\n', '')+ f' {self.tipos[indice_para_size_tipos]}{self.tamanhos[indice_para_size_tipos]}')
                lista_sql_insert.append('\t'+elemento)
              else:
                lista_sql_create.append('\t'+elemento.replace('\n', '')+ f' {self.tipos[indice_para_size_tipos]}({self.tamanhos[indice_para_size_tipos]})')
                lista_sql_insert.append('\t'+elemento)

            else:
              if self.tamanhos[indice_para_size_tipos] == '':
                lista_sql_create.append('\t'+elemento.replace('\n', '')+ f' {self.tipos[indice_para_size_tipos]}{self.tamanhos[indice_para_size_tipos]}'+',\n')
                lista_sql_insert.append('\t'+elemento+','+'\n')
              else:
                lista_sql_create.append('\t'+elemento.replace('\n', '')+ f' {self.tipos[indice_para_size_tipos]}({self.tamanhos[indice_para_size_tipos]})'+',\n')
                lista_sql_insert.append('\t'+elemento+','+'\n')
            indice_para_size_tipos += 1

          lista_sql_create.append(');\n')
          lista_sql_insert.append(') VALUES ')
        else:
          lista_sql_insert.append('(')
          for dado in linha.split(','):
            if '"' in dado:
              dado = dado.replace('"', "'")
              lista_sql_insert.append(f"{dado}"+',')
            elif '.' in dado and '-' in dado:
              dado = float(dado)
              lista_sql_insert.append(f"{dado}"+',')
            else:
              lista_sql_insert.append(f"'{dado}'"+',')
          lista_sql_insert.append('),')
      sql_create = ''.join(lista_sql_create)
      sql_insert = ''.join(lista_sql_insert).replace(',)', ')')[:-1]
      sql_create.replace('"', '')
      sql_create.replace('"', "'")
      sql_insert.replace('"', '')
      sql_insert.replace('"', "'")
      return sql_create.replace('"', "")+sql_insert.replace('"', "")

      

          
      def escrever_e_salvar(self):
        arquivo_sql = open(f'{self.nome_arquivo}.sql', 'w')
        arquivo_sql.write(sql_create+sql_insert)
        arquivo_sql.write(';')
        arquivo_sql.close()
      
teste = SqlToCsv()
arquivo_teste = teste.to_sql(caminho_arquivo_csv='arquivo_teste.csv',
             tipos=['INT','DOUBLE PRECISION', 'DOUBLE PRECISION', 'VARCHAR', 'VARCHAR'], 
             tamanhos=[10,0,0,50,2],
             nome_arquivo = 'olist_geolocation_dataset')
contar_tempo()
```
