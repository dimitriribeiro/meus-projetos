# Função - Produzindo Gráficos DRX




````python
dicionario_drx = {'Amostras':['H-1AMB~1.TXT','H-2AMB~1.TXT','H-2AMB~2.TXT','H-3AMB~1.TXT','H-9AMB~1.TXT','H-18AM~1.TXT'], 
              'Branco':['H-1BRA~1.TXT','H-2BRA~1.TXT','H-3BRA~1.TXT','H-9BRA~1.TXT','H-18BR~1.TXT']}
def plot_drx(arquivo, inicio, caminho):
    #Tratamento dos dados
    df = pd.read_csv(r'C:\Users\LASID\Desktop\Yuri\fwdfwresultadosdrx20221129\{}'.format(arquivo), 
                     sep=',')
    df = df[inicio:]
    df[df.columns[0]] = df[df.columns[0]].apply(lambda x: float(x))
    df[df.columns[1]] = df[df.columns[1]].apply(lambda x: float(x))
    #Produzindo o gráfico
    plt.figure(figsize=(10,8))
    plt.title('{}-{}'.format('DRX',df.columns[1]))
    sns.set('paper')
    ax = sns.lineplot(x = df[df.columns[0]], y = df[df.columns[1]], color='black')
    plt.savefig(caminho + "\\drx-{}.png".format(df.columns[1]), dpi=300)
````

Rodando função

````python
for amostra in dicionario['Amostras']:
    plot_drx(amostra, 14, r"C:\Users\LASID\Desktop\Yuri\DRX - Fabia")
for amostra in dicionario['Branco']:
    plot_drx(amostra, 14, r"C:\Users\LASID\Desktop\Yuri\DRX - Fabia")
````

# Função - Produção de Gráficos FTIR

````python
def plot_ftir(arquivo, inicio, caminho):
    #Tratamento dos dados
    df = pd.read_csv(r'C:\Users\LASID\Desktop\Yuri\fwdfwftirufffilmesamb\{}'.format(arquivo), 
                     sep=',')
    df = df[inicio:]
    
    if 'amb' in arquivo.lower():
        df.columns = ['Amostra', "{}-{}".format(arquivo[0:3], 'AmB')]
    else:
        df.columns = ['Amostra', "{}-{}".format(arquivo[0:3], 'BR')]
    
    df[df.columns[0]] = df[df.columns[0]].apply(lambda x: float(x))
    df[df.columns[1]] = df[df.columns[1]].apply(lambda x: float(x))
    #Produzindo o gráfico
    plt.figure(figsize=(10,8))
    plt.title('{}-{}'.format('FTIR',df.columns[1]))
    sns.set('paper')
    ax = sns.lineplot(x = df[df.columns[0]], y = df[df.columns[1]], color='black')
    plt.savefig(caminho + "\\ftir-{}.png".format(df.columns[1]), dpi=300)
````
